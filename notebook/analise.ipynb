```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/vendas.csv')

df['faturamento'] = df ['quantidade'] * df['preco']

df.head()
```

```python
faturamento_total = df['faturamento'].sum()
print(f'Faturamento total: R${faturamento_total}')
```

```python
mais_vendido = df.groupby('produto')['quantidade'].sum().sort_values(ascending=False)
print(mais_vendido)
```

```python
produto_regiao = df.groupby(['regiao', 'produto'])['quantidade'].sum()
print(produto_regiao)
```

```python
vendas_regiao = df.groupby('regiao')['faturamento'].sum()

vendas_regiao.plot(kind='bar')
plt.title('Faturamento por Região')
plt.xlabel('Região')
plt.ylabel('Faturamento')
plt.show()
plt.savefig('../img/faturamento_região.png')
```

# Análise de Desempenho e Insights
Principais Insights:
- Liderança Regional: A região Sudeste consolidou-se como o mercado mais relevante, registrando o maior volume de transações (27 vendas) e atingindo um faturamento de R$ 15.300.

- Mix de Produtos: O segmento de Celulares é o principal driver de receita do portfólio, com uma contribuição absoluta de R$ 16.000, apresentando o maior impacto direto no faturamento total.

- Oportunidade de Mercado: A região Nordeste apresenta a menor representatividade financeira atual, com um faturamento de R$ 8.900, indicando um potencial de expansão ou necessidade de revisão estratégica.

# Recomendações Estratégicas
- Manutenção de Market Share: Intensificar as campanhas de marketing na região Sudeste para capitalizar sobre a alta conversão de vendas já existente e fidelizar a base de clientes.

- Foco no Core Business: Priorizar investimentos em estoque e ações promocionais para a linha de Celulares, dado que este produto detém a maior participação na receita (share of wallet).

- Expansão Regional: Realizar um estudo de viabilidade para novas campanhas direcionadas à região Nordeste. O objetivo é mitigar a baixa performance atual e ampliar a penetração da marca em mercados menos explorados.
